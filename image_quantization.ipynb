```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread("./1.jpeg")
shape = img.shape
img = img[:,:,0]
print("plz input number of gray scale level:")
num_thresh = int(input())
img_copy = np.zeros(img.shape,np.uint8)
slice = np.linspace(0,255,num_thresh - 1)

for i in range(shape[0]):
    for j in range(shape[1]):
        for k in range(len(slice)):
            if img[i,j] > slice[k] and img[i,j]<= slice[k+1]:
                img_copy[i,j] = k
                      
cv2_imshow(img_copy)


```
