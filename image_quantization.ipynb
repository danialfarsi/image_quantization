```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread("./1.jpeg")
shape = img.shape
img = img[:,:,0]
num_thresh = int(input("plz input number of gray scale level:"))
if num_thresh < 2:
    print("Invalid number of levels!")
    exit()
img_copy = np.zeros(img.shape,np.uint8)
unit_size = 256 // num_thresh
slice = np.arange(0, 256, unit_size)
for i in range(num_thresh):
    img_copy[(img >= slice[i]) & (img < slice[i+1])] = i * unit_size 
cv2_imshow(img_copy)


```
